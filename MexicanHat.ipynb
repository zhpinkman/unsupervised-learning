### Importing the required libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
input_arr = np.array([.32, .33, .28, .47, .66, .8, .4, .33, .1, .26])
n = input_arr.shape[0]
```

```python
def activate(x):
    if x < 2:
        return 0
    elif 0 <= x < 2:
        return x
    else:
        return 2
```

```python
def process_inputs(C1, C2):
    x_old = input_arr.copy()
    x = np.zeros_like(x_old)
    results = []
    index = 0
    t_max = 5
    while index < t_max:
        index += 1
        for i in range(n):
            x[i] = 0
            for j in range(max(0, i - R1), min(n, i + R1 + 1)):
                x[i] += x_old[j]*C1
            for j in range(min(i + R1 + 1, n), min(i + R2 + 1, n)):
                x[i] += x_old[j]*C2
            for j in range(max(0, i - R2), max(0, i - R1)):
                x[i] += x_old[j]*C2
            x[i] = activate(x[i])
        x_old = x.copy()
        results.append(x_old)
        print(x_old)
    return results
```

### Processing the array using Mexican Hat model using R1 = 0 and R2 = np.inf

```python
R1 = 0
R2 = np.inf
```

```python
C1 = 3
C2 = -.1
results = process_inputs(C1, C2)
for result in results:
    plt.plot(result)
plt.title('the ranks of the number of input array capturing just the maximum number')
plt.yticks([])
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)
plt.xticks(np.arange(0, 10), np.arange(1, 11))
plt.xlabel('index of the numbers in the array')
plt.show()
```

```python
C1 = 4
C2 = -.1
results = process_inputs(C1, C2)
for result in results:
    plt.plot(result)
plt.title('the ranks of the number of input array capturing the maximum number and the second largets number')
plt.yticks([])
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)
plt.xticks(np.arange(0, 10), np.arange(1, 11))
plt.xlabel('index of the numbers in the array')
plt.show()
```
